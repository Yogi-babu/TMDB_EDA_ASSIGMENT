```python
import requests
import pandas as pd
import sqlite3
import time
import json

API_KEY = "3155e101354aec2537082ac90b3aee9e"

movies = []
url = "https://api.themoviedb.org/3/discover/movie"

headers = {
    "User-Agent": "Mozilla/5.0"
}

for page in range(1, 3):
    params = {
        "api_key": API_KEY,
        "page": page
    }
    
    try:
        response = requests.get(url, params=params, headers=headers, timeout=10)
        response.raise_for_status()
        
        data = response.json()
        
        if 'results' in data:
            movies.extend(data['results'])
        
        time.sleep(1)
        
    except Exception as e:
        print("Error:", e)

if len(movies) == 0:
    print(" No data fetched. Check API key or network.")

else:
    df = pd.DataFrame(movies)
    
    if 'genre_ids' in df.columns:
        df['genre_ids'] = df['genre_ids'].apply(lambda x: json.dumps(x) if isinstance(x, list) else x)
    
    conn = sqlite3.connect("movies.db")
    df.to_sql("movies", conn, if_exists="replace", index=False)
    
    df = pd.read_sql("SELECT * FROM movies", conn)
    conn.close()
    
    
    print("\n DATA FETCHED SUCCESSFULLY")
    
    print("\n Shape of Data:")
    display(pd.DataFrame({"Rows": [df.shape[0]], "Columns": [df.shape[1]]}))
    
    print("\n First 5 Rows:")
    display(df.head())
    
    print("\n Summary Statistics:")
    display(df.describe())
    
    print("\n Missing Values:")
    display(df.isnull().sum().to_frame(name="Missing Count"))
    
    print("\n Top Genre Counts:")
    display(df['genre_ids'].value_counts().head().to_frame(name="Count"))
```
